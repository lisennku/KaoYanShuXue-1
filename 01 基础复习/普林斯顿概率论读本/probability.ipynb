```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
```

```python
trial = [n for n in range(2, 51)]
result = [0] * len(trial)
init_prob = 1.0
for i,n in enumerate(trial):
    prob = (365-(n-1)) / 365
    init_prob *= prob
    result[i] = 1 - init_prob
```

```python
# 绘图
plt.figure(figsize=(10, 6))  # 设置图形大小
plt.plot(trial, result, 'o-', color='blue', markersize=6, label='碰撞概率')  # 添加线条和标记

# 添加标题和标签
plt.title('生日问题碰撞概率', fontsize=16)
plt.xlabel('人数 (n)', fontsize=14)
plt.ylabel('碰撞概率', fontsize=14)

# 添加网格
plt.grid(True, linestyle='--', alpha=0.7)

# 设置轴范围
plt.xlim(1, 51)
plt.ylim(0, 1)

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
