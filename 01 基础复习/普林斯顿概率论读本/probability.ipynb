```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
```

```python
trial = [n for n in range(2, 51)]
result = [0] * len(trial)
init_prob = 1.0
for i,n in enumerate(trial):
    prob = (365-(n-1)) / 365
    init_prob *= prob
    result[i] = 1 - init_prob
```

```python
# 
plt.figure(figsize=(10, 6))  # 
plt.plot(trial, result, 'o-', color='blue', markersize=6, label='')  # 

# 
plt.title('', fontsize=16)
plt.xlabel(' (n)', fontsize=14)
plt.ylabel('', fontsize=14)

# 
plt.grid(True, linestyle='--', alpha=0.7)

# 
plt.xlim(1, 51)
plt.ylim(0, 1)

# 
plt.minorticks_on()  # 
plt.gca().xaxis.set_minor_locator(plt.MultipleLocator(1))  # 1

# 
plt.legend()

# 
plt.show()
```
